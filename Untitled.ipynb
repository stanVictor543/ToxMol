```python
#Importuri
import deepchem as dc
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem.AllChem import GetMorganFingerprintAsBitVect
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score, precision_recall_curve
import time

print("--- Inițializare Proiect: Previziunea Toxicității NR-AR ---")

# Parametri globali
TARGET_ASSAY = 'NR-AR'
N_BITS = 2048 
RADIUS = 2
```

```python
#  Încărcarea setului de date Tox21
try:
    print(f"Încărcarea datelor Tox21... (Țintă: {TARGET_ASSAY})")
    tasks, datasets, transformers = dc.molnet.load_tox21()
    train_dataset, valid_dataset, test_dataset = datasets[0], datasets[1], datasets[2]
    
    # Găsește indexul coloanei țintă
    atad5_index = tasks.index(TARGET_ASSAY)
    print(f"Date încărcate cu succes. Indexul pentru '{TARGET_ASSAY}' este {atad5_index}.")
    
except Exception as e:
    print(f"Eroare: {e}")
```

```python
#  Funcții Utilitare
def smiles_to_morgan_fingerprint(smiles, nBits=N_BITS, radius=RADIUS):
    """Convertește șirul SMILES în Morgan Fingerprint."""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None 
        fingerprint = GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
        return np.array(fingerprint)
    except:
        return None

def prepare_data(dataset, target_index):
    """Extrage, filtrează și generează descriptorii pentru un set de date."""
    print("   ... Procesare set de date ...")
    X_smiles = dataset.ids
    Y_all = dataset.y
    Y_target = Y_all[:, target_index]

    # Filtrare NaN
    nan_mask_y = ~np.isnan(Y_target)
    X_smiles_cleaned = X_smiles[nan_mask_y]
    Y_cleaned = Y_target[nan_mask_y].astype(int)

    # Generarea Fingerprints
    X_fingerprints = [smiles_to_morgan_fingerprint(s) for s in X_smiles_cleaned]

    # Filtrare molecule invalide
    valid_indices = [i for i, fp in enumerate(X_fingerprints) if fp is not None]
    X_final = np.array([X_fingerprints[i] for i in valid_indices])
    Y_final = Y_cleaned[valid_indices].ravel()

    print(f"   -> Dimensiune Finală (X, Y): {X_final.shape}")
    return X_final, Y_final

def evaluate_model(model, X_test, Y_test, model_name):
    """Calculează metricile de bază."""
    Y_pred = model.predict(X_test)
    Y_proba = model.predict_proba(X_test)[:, 1]

    auc_roc = roc_auc_score(Y_test, Y_proba)
    f1 = f1_score(Y_test, Y_pred)
    accuracy = accuracy_score(Y_test, Y_pred)

    print(f"\n   --- Evaluare: {model_name} ---")
    print(f"   - AUC-ROC: {auc_roc:.4f}")
    print(f"   - F1-Score (Standard): {f1:.4f}")
    print(f"   - Acuratețe: {accuracy:.4f}")
    return auc_roc, f1
```

```python
#  Generarea Fingerprints (Poate dura 1-2 minute)
print("\nPregătirea Datelor (Generare Fingerprints)...")

print("Set Antrenament:")
X_train, Y_train = prepare_data(train_dataset, atad5_index)

print("Set Test:")
X_test, Y_test = prepare_data(test_dataset, atad5_index)
```

```python
#  Bayesian Naiv
print("\n4.1 Antrenare Model 1: Clasificator Bayesian Naiv...")
start_time_bayes = time.time()

model_bayes = GaussianNB()
model_bayes.fit(X_train, Y_train)

time_bayes = time.time() - start_time_bayes
print(f"   -> Timp antrenament: {time_bayes:.2f} secunde")

auc_bayes, f1_bayes = evaluate_model(model_bayes, X_test, Y_test, "Bayesian Naiv")
```

```python
# Random Forest
print("\n4.2 Antrenare Model 2: Random Forest...")
start_time_rf = time.time()

# Folosim class_weight='balanced'
model_rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1) 
model_rf.fit(X_train, Y_train)

time_rf = time.time() - start_time_rf
print(f"   -> Timp antrenament: {time_rf:.2f} secunde")

auc_rf, f1_rf = evaluate_model(model_rf, X_test, Y_test, "Random Forest")
```

```python
#  Tabel Rezultate și Feature Importance
results = {
    'Algoritm': ['Bayesian Naiv', 'Random Forest'],
    'AUC-ROC': [auc_bayes, auc_rf],
    'F1-Score (Default)': [f1_bayes, f1_rf],
    'Timp (s)': [time_bayes, time_rf]
}
df_results = pd.DataFrame(results)
print("REZULTATE FINALE:")
print(df_results.sort_values(by='AUC-ROC', ascending=False).to_markdown(index=False))

# Feature Importance pentru RF
print("\nTop 10 Biți Importanți (Structuri Chimice) în Random Forest:")
importances = model_rf.feature_importances_
feature_names = [f"Bit_{i}" for i in range(N_BITS)]
feature_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
print(feature_df.sort_values(by='Importance', ascending=False).head(10).to_markdown(index=False))
```

```python
#  Optimizarea Pragului de Decizie
print("--- Optimizarea Pragului de Decizie (Precision-Recall Curve) ---")

Y_proba_rf = model_rf.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(Y_test, Y_proba_rf)

# Calcul F1 pentru toate pragurile posibile
fscore = 2 * (precision * recall) / (precision + recall)
fscore[np.isnan(fscore)] = 0 

# Găsirea optimului
ix = np.argmax(fscore)
optimal_threshold = thresholds[ix]
f1_optimized = fscore[ix]

print(f"F1-Score Maximizat: {f1_optimized:.4f}")
print(f"Prag Optim: {optimal_threshold:.4f}")

# Re-evaluare finală
Y_pred_optimized = (Y_proba_rf >= optimal_threshold).astype(int)
print(f"\nPerformanță Finală Random Forest (după optimizare):")
print(f"AUC-ROC: {roc_auc_score(Y_test, Y_proba_rf):.4f}")
print(f"F1-Score: {f1_score(Y_test, Y_pred_optimized):.4f}")
print(f"Acuratețe: {accuracy_score(Y_test, Y_pred_optimized):.4f}")
```

```python
# Celula 10: Vizualizare Grafică (ROC și Precision-Recall)
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve, auc

def plot_performance(y_true, y_proba, optimal_thresh):
    plt.figure(figsize=(14, 6))

    # --- Grafic 1: Curba ROC ---
    plt.subplot(1, 2, 1)
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.3f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--') # Linia de ghicire aleatorie
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (Alarme False)')
    plt.ylabel('True Positive Rate (Sensibilitate)')
    plt.title('Curba ROC')
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)

    # --- Grafic 2: Precision-Recall Curve ---
    plt.subplot(1, 2, 2)
    precision, recall, thresholds = precision_recall_curve(y_true, y_proba)
    
    plt.plot(recall, precision, color='green', lw=2, label='Precision-Recall curve')
    
    # Marcăm punctul optim pe grafic
    # (Trebuie să găsim indexul celui mai apropiat prag de cel optim calculat anterior)
    closest_idx = np.argmin(np.abs(thresholds - optimal_thresh))
    plt.scatter(recall[closest_idx], precision[closest_idx], s=100, c='red', label=f'Prag Optim ({optimal_thresh:.2f})')
    
    plt.xlabel('Recall (Câtă toxicitate am prins?)')
    plt.ylabel('Precision (Cât de sigur e modelul?)')
    plt.title('Curba Precision-Recall (Critică pentru date dezechilibrate)')
    plt.legend(loc="best")
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

# Apelăm funcția folosind datele de la Random Forest
plot_performance(Y_test, Y_proba_rf, optimal_threshold)
```
